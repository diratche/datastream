````python
from __future__ import print_function
import time
import weatherapi
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from weatherapi.rest import ApiException
from pprint import pprint
from dotenv import load_dotenv
import os

load_dotenv()
````

````python
# Configure API key authorization: ApiKeyAuth
configuration = weatherapi.Configuration()
configuration.api_key["key"] = os.getenv("weather_api_key")
# Uncomment below to setup prefix (e.g. Bearer) for API key, if needed
# configuration.api_key_prefix['key'] = 'Bearer'


# create an instance of the API class
api_instance = weatherapi.APIsApi(weatherapi.ApiClient(configuration))
q = "Kherson"  # str | Pass US Zipcode, UK Postcode, Canada Postalcode, IP address, Latitude/Longitude (decimal degree) or city name. Visit [request parameter section](https://www.weatherapi.com/docs/#intro-request) to learn more.
dt = "2024-02-20"  # date | Date on or after 1st Jan, 2015 in yyyy-MM-dd format
unixdt = 56  # int | Please either pass 'dt' or 'unixdt' and not both in same request.<br />unixdt should be on or after 1st Jan, 2015 in Unix format (optional)
end_dt = "2024-02-30"  # date | Date on or after 1st Jan, 2015 in yyyy-MM-dd format<br />'end_dt' should be greater than 'dt' parameter and difference should not be more than 30 days between the two dates. (optional)
unixend_dt = 56  # int | Date on or after 1st Jan, 2015 in Unix Timestamp format<br />unixend_dt has same restriction as 'end_dt' parameter. Please either pass 'end_dt' or 'unixend_dt' and not both in same request. e.g. unixend_dt=1490227200 (optional)
# hour = 56 # int | Must be in 24 hour. For example 5 pm should be hour=17, 6 am as hour=6 (optional)
lang = "en"  # str | Returns 'condition:text' field in API in the desired language.<br /> Visit [request parameter section](https://www.weatherapi.com/docs/#intro-request) to check 'lang-code'. (optional)

try:
    # History API
    api_response = api_instance.history_weather(q, dt, end_dt=end_dt, lang=lang)
    pprint(api_response)
except ApiException as e:
    print("Exception when calling APIsApi->history_weather: %s\n" % e)
````

````python
# Key between the region name in ACLED and the name used in the weather API
region_spell_ACLED2weather = {
    "Zaporizhia": "Zaporizhzhya",
    "Mykolaiv": "Mykolayiv",
    "Vinnytsia": "Vinnytsya",
}
````

````python
# check the valid regions for the API

dt = "2024-12-01"
all_responses = []
for region in regions:
    q = (
        region
        if region not in region_spell_ACLED2weather
        else region_spell_ACLED2weather[region]
    )
    try:
        # History API
        api_response = api_instance.history_weather(q, dt, end_dt="2024-12-05")
        if api_response["location"]["country"] not in ["Ukraine", "Russia"]:
            continue
        # pprint(api_response)
        day_response = pd.DataFrame(
            api_response["forecast"]["forecastday"][0]["hour"]
        )
        day_response[["region", "lat", "lon"]] = (
            api_response["location"]["name"],
            api_response["location"]["lat"],
            api_response["location"]["lon"],
        )
        all_responses.append(day_response)
    except ApiException as e:
        print(f"Exception when calling history_weather for {q}: {e}")
all_responses = pd.concat(all_responses)
all_responses["time"] = pd.to_datetime(all_responses["time"])
all_responses = all_responses[["region", "lat", "lon"] + selected_hour_keys]
all_responses.set_index("time", inplace=True)
all_responses.head()
````

## Load all data for last year in area of fight

````python
all_responses = []
for day in pd.date_range(start="2023-12-01", end="2024-12-05"):
    dt = day.strftime("%Y-%m-%d")

    for region in regions:
        q = (
            region
            if region not in region_spell_ACLED2weather
            else region_spell_ACLED2weather[region]
        )
        try:
            # History API
            api_response = api_instance.history_weather(
                q,
                dt,
            )
            if api_response["location"]["country"] not in ["Ukraine", "Russia"]:
                continue
            # pprint(api_response)
            day_response = pd.DataFrame(
                api_response["forecast"]["forecastday"][0]["hour"]
            )
            day_response[["region", "lat", "lon"]] = (
                api_response["location"]["name"],
                api_response["location"]["lat"],
                api_response["location"]["lon"],
            )
            all_responses.append(day_response)
        except ApiException as e:
            print(f"Exception when calling history_weather for {q}/{dt}: {e}")
all_responses = pd.concat(all_responses)
all_responses["time"] = pd.to_datetime(all_responses["time"])
all_responses = all_responses[["region", "lat", "lon"] + selected_hour_keys]
all_responses.set_index("time", inplace=True)
all_responses.head()
````
